(1):A Simple Linear Regression model describes the relationship between two variables by modeling the outcome variable y as a linear function of the predictor variable x.The model equation y =β0 +β1x + ε includes an intercept β0, which is the starting value of y when x is zero, and a slope β1, which shows the rate of change in y for each one-unit change in x. The error term ε accounts for variability in y that isn’t explained by x, assumed to be normally distributed with a mean of zero and constant variance. This normal distribution assumption means observed y values are clustered around the predicted values (the regression line), forming a bell curve around the line for each x, making it possible to interpret and predict outcomes based on the linear relationship between x and y.

```python
import numpy as np
from scipy.stats import uniform, norm
import plotly.graph_objects as go

# Model parameters
n = 100          # Number of data points
beta_0 = 2       # Intercept
beta_1 = 0.8     # Slope
sigma = 1.0      # Standard deviation of the error term

# Step 1: Generate predictor values (X) using a uniform distribution
X = uniform.rvs(0, 10, size=n)

# Step 2: Generate error terms (epsilon) using a normal distribution
epsilon = norm.rvs(0, sigma, size=n)

# Step 3: Calculate outcome variable (Y) based on the SLR model
Y = beta_0 + beta_1 * X + epsilon

# Visualization
fig = go.Figure()

# Scatter plot of the generated data points
fig.add_trace(go.Scatter(x=X, y=Y, mode='markers', name='Simulated Data'))

# Line plot of the theoretical relationship without error
line_Y = beta_0 + beta_1 * X
fig.add_trace(go.Scatter(x=X, y=line_Y, mode='lines', name='Theoretical Line'))

# Customize layout
fig.update_layout(
    title="Simple Linear Regression Model Simulation",
    xaxis_title="Predictor Variable X",
    yaxis_title="Outcome Variable Y"
)

fig.show()
```

Theoretical Line(from Q1): This line represents the true underlying relationship based on specified parameters(β 0and β1) and is plotted as an orange dashed line. It’s derived from our initial theoretical model equation Y=β0+β1X without any random error component.

Fitted Regression Line (from Q2): This line is estimated from the simulated dataset. The line is derived using ordinary least squares (OLS) on the simulated data, so it represents the best-fit line based on the sample data rather than the exact theoretical model.

The theoretical line represents the “true” model relationship as specified by fixed values of β0 and β1. In contrast, the fitted line accounts for random sampling variation in the data due to the error term ϵ in the model.

Because of this random noise, the fitted line often deviates slightly from the true theoretical line. The difference represents the inherent sampling variability in real-world data, where each sample might lead to a slightly different fitted line.

In a Simple Linear Regression model, we have the following theoretical equation:
Yi =β0 +β1xi +ϵi 

Yi is the observed outcome for each i-th observation.
xi is thepredictorvariable for each i-th observation.
β0(intercept) and β1(slope) are the true parameters of the model.
ϵi is the error term, assumed to follow a normal distribution N(0,σ).
However, in practice, we don’t know the true values of β0 and β1. Instead, we estimate them using the sample data. When we fit the model, we obtain estimates ^β0 and ^β1, which are stored in fitted_model.params.

The fitted (predicted) values, fitted_model.fittedvalues, are the in-sample predictions made using the estimated model. They are calculated as follows:^Yi = ^β0 + ^β1xi 

^Yi is the predicted (fitted) value for the i-th observation.
^β0 and ^β1 are the estimated intercept and slope from the fitted model, which can be accessed through fitted_model.params.


（5）：
OLS method finds the best-fitting line by minimizing the sum of the squared differences between the observed Y values and the predicted ^Y values on this line.The OLS method chooses the line that minimizes the sum of the squared residuals. The reason for squaring the residuals is to avoid negative values canceling out positive ones, ensuring that all deviations are treated as positive contributions to the error. This approach emphasizes larger residuals more than smaller ones, penalizing larger errors and leading to a line that best represents the central trend of the observed data.

Why OLS Requires "Squares":
Emphasizes Larger Deviations: Squaring the residuals penalizes large deviations more than small ones, which helps the model avoid large errors and better fit the majority of the data points.
Prevents Cancellation: If residuals were summed directly, positive and negative errors could cancel each other out, which would not accurately represent the model’s fit. Squaring removes this problem, allowing OLS to find a line that minimizes the total error in a consistent way.

An R^2 value of 0 means the model explains none of the variation in Y, and the predictions are no better than simply using the mean of Y for all predictions.
Thus, fitted_model.rsquared serves as a measure of how accurately the model describes the relationship between X and Y. A higher R^2 value indicates a better fit to the data.

```python
#（10）
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns

old_faithful = sns.load_dataset('geyser')
long_wait_limit = 71
long_wait = old_faithful.waiting > long_wait_limit

# Step 1: Bootstrapped sampling for fitted slope coefficients
bootstrapped_slope_coefficients = []
n_iterations = 1000

for _ in range(n_iterations):
    sample = old_faithful[long_wait].sample(frac=1, replace=True)
    model = smf.ols('duration ~ waiting', data=sample).fit()
    bootstrapped_slope_coefficients.append(model.params['waiting'])

# Step 2: Simulated sampling under null hypothesis (no linear association)
simulated_slope_coefficients = []
n_simulations = 1000

# Using beta0 = 1.65, beta1 = 0, sigma = 0.37
beta0, beta1, sigma = 1.65, 0, 0.37
old_faithful_simulation = old_faithful[long_wait].copy()

for _ in range(n_simulations):
    old_faithful_simulation['duration'] = beta0 + beta1 * old_faithful_simulation['waiting'] + \
                                          np.random.normal(0, sigma, len(old_faithful_simulation))
    model = smf.ols('duration ~ waiting', data=old_faithful_simulation).fit()
    simulated_slope_coefficients.append(model.params['waiting'])

# Step 3: Check if 0 is within the 95% confidence interval of the bootstrapped slopes
ci_lower, ci_upper = np.quantile(bootstrapped_slope_coefficients, [0.025, 0.975])
contains_zero = ci_lower <= 0 <= ci_upper

print(f"95% CI for bootstrapped slope coefficients: [{ci_lower}, {ci_upper}]")
print(f"Contains zero: {contains_zero}")

# Calculate p-value for simulated slope distribution
p_value = (np.abs(simulated_slope_coefficients) >= np.abs(np.mean(bootstrapped_slope_coefficients))).mean()
print(f"Simulated p-value: {p_value}")
```

```python
#（12）：
from plotly.subplots import make_subplots
import plotly.graph_objects as go
from scipy import stats
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
import pandas as pd

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Define a new categorical variable 'kind' based on wait times
old_faithful['kind'] = np.where(old_faithful['waiting'] < 68, 'short', 'long')

# Model 1: Using all data with waiting time as a continuous predictor
model1 = smf.ols('duration ~ waiting', data=old_faithful).fit()
residuals1 = model1.resid

# Model 2: Using only short wait times
short_wait = old_faithful[old_faithful['waiting'] < 64]
model2 = smf.ols('duration ~ waiting', data=short_wait).fit()
residuals2 = model2.resid

# Model 3: Using only long wait times
long_wait = old_faithful[old_faithful['waiting'] > 71]
model3 = smf.ols('duration ~ waiting', data=long_wait).fit()
residuals3 = model3.resid

# Model 4: Using all data with an indicator variable 'kind'
model4 = smf.ols('duration ~ C(kind, Treatment(reference="short"))', data=old_faithful).fit()
residuals4 = model4.resid

# Dictionary to store residuals for each model
model_residuals = {
    'Model 1:<br>All Data using Slope': residuals1,
    'Model 2:<br>Short Wait Data': residuals2,
    'Model 3:<br>Long Wait Data': residuals3,
    'Model 4:<br>All Data using Indicator': residuals4
}

# Create subplots
fig = make_subplots(rows=2, cols=2, subplot_titles=list(model_residuals.keys()))

# Plot histograms with normal distribution curve
for i, (title, resid) in enumerate(model_residuals.items()):
    # Compute normal range for comparison line
    normal_range = np.linspace(resid.min(), resid.max(), 100)
    normal_dist = stats.norm(loc=resid.mean(), scale=resid.std()).pdf(normal_range)
    
    # Determine subplot position
    row = int(i / 2) + 1
    col = (i % 2) + 1
    
    # Add histogram
    fig.add_trace(go.Histogram(x=resid, name=title, histnorm='probability density', 
                               marker=dict(color='skyblue'), nbinsx=15), row=row, col=col)
    
    # Add normal distribution line
    fig.add_trace(go.Scatter(x=normal_range, y=normal_dist, mode='lines', 
                             name="Normal Distribution", line=dict(color='black', dash='dash')), 
                  row=row, col=col)

# Update layout
fig.update_layout(title_text="Histograms of Residuals from Different Models", showlegend=False,
                  xaxis_title="Residuals", yaxis_title="Density")

fig.show()
```

```python
#（13）：
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf

# Load the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')

# Create a new "kind" column, categorizing waiting times <68 as "short" and >68 as "long"
old_faithful['kind'] = np.where(old_faithful['waiting'] < 68, 'short', 'long')

# (A) Permutation Test
np.random.seed(0)
observed_mean_diff = old_faithful.groupby('kind')['duration'].mean().diff().iloc[-1]

permuted_diffs = []
for _ in range(1000):  # Repeat 1000 times
    # Randomly shuffle labels
    shuffled = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1).values)
    # Calculate the mean difference between shuffled groups
    mean_diff = shuffled.groupby('kind_shuffled')['duration'].mean().diff().iloc[-1]
    permuted_diffs.append(mean_diff)

# Calculate the p-value for the permutation test
permuted_diffs = np.array(permuted_diffs)
p_value = np.mean(np.abs(permuted_diffs) >= np.abs(observed_mean_diff))
print(f"(A) Permutation test p-value: {p_value:.4f}")

# (B) Bootstrap Confidence Interval
bootstrapped_mean_diffs = []
for _ in range(1000):  # 1000 resamples
    # Bootstrap sampling for each group
    short_sample = old_faithful[old_faithful['kind'] == 'short']['duration'].sample(frac=1, replace=True)
    long_sample = old_faithful[old_faithful['kind'] == 'long']['duration'].sample(frac=1, replace=True)
    # Calculate mean difference
    mean_diff = long_sample.mean() - short_sample.mean()
    bootstrapped_mean_diffs.append(mean_diff)

# Calculate the 95% Bootstrap Confidence Interval
ci_lower, ci_upper = np.percentile(bootstrapped_mean_diffs, [2.5, 97.5])
print(f"(B) 95% Bootstrap Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
```

(a)：
In the permutation test, we shuffle the labels to simulate the null distribution under the assumption that there is no difference between groups. This tests if the observed difference in means is likely to occur by random chance.
In bootstrapping, we repeatedly sample from the original groups to create a distribution of possible mean differences, estimating the variability of the observed difference.
(b)：
Similarity: Both the permutation test and bootstrapping are resampling methods that do not assume a specific distribution, making them flexible and useful for creating confidence intervals or testing hypotheses.
Difference: The indicator variable model in Question 11 is based on a regression approach, where a categorical indicator variable separates the 'short' and 'long' wait times directly within a model. This approach provides an estimate of the effect of group (short vs. long) within a linear model context, rather than purely testing the difference between groups as done in permutation and bootstrapping.
